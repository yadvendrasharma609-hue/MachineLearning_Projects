```python
# ============================================
# Machine Learning I Lab
# Iris Dataset Analysis using NumPy, Pandas, and Matplotlib
# ============================================

# Importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ============================================
# 1. Load the Iris dataset
# (Seaborn has it built-in; otherwise we can load from CSV)
# ============================================
iris = sns.load_dataset('iris')

# ============================================
# 2. Preview the dataset
# ============================================
print("\nFirst 5 rows of the Iris dataset:")
print(iris.head())

# ============================================
# 3. Dataset shape (rows, columns)
# ============================================
print(f"\nDataset shape: {iris.shape} (rows, columns)")

# ============================================
# 4. Column names
# ============================================
print("\nColumn names:", iris.columns.tolist())

# ============================================
# 5. Dataset info (data types, non-null counts)
# ============================================
print("\nDataset Information:")
print(iris.info())

# ============================================
# 6. Summary statistics for numerical features
# ============================================
print("\nSummary Statistics:")
print(iris.describe())

# ============================================
# 7. Check for missing values
# ============================================
print("\nMissing values in each column:")
print(iris.isnull().sum())

# ============================================
# 8. Class distribution (species count)
# ============================================
print("\nClass distribution:")
print(iris['species'].value_counts())

# ============================================
# 9. NumPy-based calculations for each feature
# ============================================
print("\nNumPy Calculations (Mean, Median, Min, Max):")
for col in iris.columns[:-1]:  # Exclude 'species'
    col_data = iris[col].values
    print(f"\nFeature: {col}")
    print(f"  Mean   : {np.mean(col_data):.2f}")
    print(f"  Median : {np.median(col_data):.2f}")
    print(f"  Min    : {np.min(col_data):.2f}")
    print(f"  Max    : {np.max(col_data):.2f}")

# ============================================
# 10. Correlation matrix (numerical features)
# ============================================
print("\nCorrelation Matrix:")
corr_matrix = iris.iloc[:, :-1].corr()
print(corr_matrix)

# ============================================
# 11. Data Visualizations
# ============================================

# Histogram for each feature
iris.hist(edgecolor='black', linewidth=1.2, figsize=(10, 6))
plt.suptitle("Feature Histograms", fontsize=16)
plt.show()

# Pair plot (scatterplots + histograms)
sns.pairplot(iris, hue='species')
plt.suptitle("Pair Plot of Iris Features", y=1.02)
plt.show()

# Box plots for each feature by species
plt.figure(figsize=(12, 8))
for i, col in enumerate(iris.columns[:-1]):
    plt.subplot(2, 2, i+1)
    sns.boxplot(x='species', y=col, data=iris)
    plt.title(f"Boxplot of {col} by Species")
plt.tight_layout()
plt.show()

# Heatmap of correlation matrix
plt.figure(figsize=(6, 4))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
